```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('heart_attack.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)

df.drop_duplicates(inplace=True)
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['Age'], bins=20, color='blue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['Gender'].value_counts().plot(kind='bar', color='green')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.boxplot(column='Blood Pressure (mmHg)', by='Smoking Status')
plt.title('Blood Pressure by Smoking Status')
plt.suptitle('')  # Removes the automatic title
plt.xlabel('Smoking Status')
plt.ylabel('Blood Pressure (mmHg)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['Age'], df['Cholesterol (mg/dL)'], color='red')
plt.title('Age vs Cholesterol Levels')
plt.xlabel('Age')
plt.ylabel('Cholesterol (mg/dL)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['Chest Pain Type'].value_counts().plot(kind='bar', color='purple')
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,8))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['blue', 'green'])
plt.title('Gender Distribution')
plt.ylabel('')  
plt.show()
```

```python
plt.figure(figsize=(8,8))
df['Chest Pain Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, cmap='viridis')
plt.title('Chest Pain Type Distribution')
plt.ylabel('')  
plt.show()
```
